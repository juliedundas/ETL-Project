```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
# URL of page to be scraped
url = 'https://www.archives.gov/research/alic/reference/space-timeline.html'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
print(response.url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text,'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
result = soup.find('div', class_='col-sm-10')
result
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_="col-sm-10")
results
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return year of fun fact
        year = result.find('h4')
        # Identify and return price of listing
        fun_fact = result('li')
    
        # Print results only if title, price, and link are available
        if (year and fun_fact):
            print('-------------')
            print(year)
            print(fun_fact)
    except AttributeError as e:
        print(e)
```
